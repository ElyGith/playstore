```python
import pandas as pd
import numpy as np

#Bloc de configuration

pd.set_option('display.width', 200)  # Largeur totale
pd.set_option('display.max_colwidth', 50)  # Largeur par colonne

pd.set_option('display.max_columns', None)  # Afficher toutes les colonnes
```

```python
# Fichier sur lequel on va travailler 
df = pd.read_csv('../data/googleplaystore.csv',sep=',')
```

```python
df.sample(5)
```

```python
#1474 valeurs nulle pour les rating
df[df['Rating'].isnull()]
```

```python
#Suppresions des valeurs NaN car inutilisable
df = df.dropna(subset=['Rating'])
```

```python
df.isnull().sum()
```

```python
df[df['Content Rating'].isnull()]
```

```python
df[df['Current Ver'].isnull()]
```

```python
df[df['Android Ver'].isnull()]
```

```python
# Suppresion des valeurs anormales ou null
# Les autres ont peut les laisser car non nécéssaire à notre analyse 
df = df.drop(10472) 
```

```python
df["Reviews"] = df["Reviews"].astype(int)
```

```python
df.info()
```

```python
#On va normaliser Size pour pouvoir le passer en int
# k pour 10^3 et M pour 10^6
# On passe tout en octet puis on divise par 10^6 pour avoir des mégaoctet
df['Size']=df['Size'].str.replace('M','000000')
df['Size']=df['Size'].str.replace('k','000')

df['Size']=df['Size'].replace("Varies with device",np.nan)

df['Size']=df['Size'].astype('float')


```

```python
df.sample(3)
```

```python
df['Size']=df['Size']/1000000

df['Size']
```

On passe à la colonne Installs

```python
df['Installs']=df['Installs'].str.replace(',','')
df['Installs']=df['Installs'].str.replace('+','')
```

```python
df['Installs'].unique()
```

```python
df['Installs']=df['Installs'].astype('float')
```

```python
df['Price']=df['Price'].str.replace('$','')
```

```python
df['Price']=df['Price'].astype('float')
```

```python
df.sample(3)
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
df['Last Updated']
```

```python
df.to_csv("../data/2_preprocessing_data.csv", float_format="%.6f",index=False)
```
